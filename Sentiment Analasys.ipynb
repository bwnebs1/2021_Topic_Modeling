```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('Preprocessed Dataframe.csv', sep=chr(int("15")))
df
```

```python
#감성분석 (1)
```

```python
sent_postive=df['sentiment']=='positive'
sent_negative=df['sentiment']=='negative'
```

```python
df_positive=df[sent_postive].iloc[:,[0,7]]
df_negative=df[sent_negative].iloc[:,[0,7]]
```

```python
df_train=pd.concat([df_positive, df_negative])
df_train=shuffle(df_train)
```

```python
df_train
```

```python
tf_vectorizer = CountVectorizer(min_df=1, ngram_range=(1,1)) 
tf_train_features = tf_vectorizer.fit_transform(df['NounVerbAdj'])
```

```python
vocablist = [word for word, _ in sorted(tf_vectorizer.vocabulary_.items(), key=lambda x:x[1])]
```

```python
lr_tf = LogisticRegression(max_iter=10000000)
lr_tf.fit(tf_train_features, df['sentiment'])
```

```python
#(Sentiment Analysis.ipynb #감성분석 (2))
```

```python
# Get coefficients of the model 
coefficients = lr_tf.coef_.tolist()

sorted_coefficients = sorted(enumerate(coefficients[0]), key=lambda x:x[1], reverse=True)
# 학습에 사용된 각 단어마다의 coefficient (즉 weight) 값이 존재
# coefficient값이 큰 순으로 정렬 'reverse=True'

print(sorted_coefficients[:5])
# print top 20 positive words
for word, coef in sorted_coefficients[:20]:
    print('{0:} ({1:.3f})'.format(vocablist[word], coef))
# print top 20 negative words
for word, coef in sorted_coefficients[-20:]:
    print('{0:} ({1:.3f})'.format(vocablist[word], coef))
```
